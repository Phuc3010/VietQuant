# Import library

```python
import numpy as np
import pandas as pd
import math
from math import sqrt
```

# Read data

```python
data = pd.read_csv("AAA.csv")
data.head()
```

# Calcuate stock holdings and mean matched price

```python
data['Stock_holdings'] = data['Vol'].cumsum()
mean_matched_price = []
total_vol = 0
holding_stocks = data['Stock_holdings'].tolist()
volume = data['Vol'].tolist()
price = data['Price'].tolist()
price_idx = 0
for i in range(len(volume)):
    if holding_stocks[i] == 0:
        # Reset price
        total_vol = 0
        price_idx = 0
        mean_matched_price.append(0)
    else:
        if volume[i] >= 0:
            total_vol += volume[i]
            price_idx += volume[i]*price[i]
            mean_matched_price.append(price_idx/total_vol)        
        else:
            mean_matched_price.append(mean_matched_price[-1])

data['Mean_price'] = mean_matched_price
data
```

# Calculate Return, Sharp ratio and Max drawndown

```python
risk_free_ratio = 0.0
data['Return'] = ((data['Mean_price']-data['Mean_price'].shift(1))/data['Mean_price'].shift(1)).fillna(0)
return_ratio = np.mean(data['Return']-risk_free_ratio)/(sqrt(252)*np.std(data['Return']-risk_free_ratio))
print('Return ratio: ', return_ratio)
max_price = max(price[0:-2])
max_idx = data.loc[data['Price']==max_price].index[0]
min_price = min(price[max_idx:])
max_drawndown = (max_price-min_price)/max_price
print('Max drawndown: ', max_drawndown)
data
```


# Calculate Turnover and Mean PnL

```python
profit_and_loss = [price[i]-mean_matched_price[i]*holding_stocks[i] for i in range(len(price))]
mean_pnl = sum(profit_and_loss)/len(profit_and_loss)
print('Mean profit and loss: ', mean_pnl)
def determine_price(buy, sell):
    if sell == 0:
        return buy
    return min(buy, sell)

first_price, last_price = price[0], price[1]
avg_price = (first_price+last_price)/2
turnover = [determine_price(price[i], mean_matched_price[i])/avg_price for i in range(len(price))]
data['Turnover'] = turnover
data
```
